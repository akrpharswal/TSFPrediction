In this regression problem, our goal is to estimate the expected student's score percentage by considering the number of hours they have dedicated to studying. This task can be categorized as a straightforward linear regression problem, given that it revolves around only two variables.

```python
# Importing  libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
```

```python
#Reading data
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported from Link")
#View data
data.head()
```

```python
# Visualizing the score distribution on a graph.Below scatter we see corelation between score vs hours
data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('The Hours Studied')  
plt.ylabel('The % Score')  
plt.show()
```

```python
#Attribute and labels
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values  
```

```python
#doing Random split, we use train_test_split() 
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

```python
#Training Algorithm
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training completed...")
```

```python
# implementing the plotting test data 
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

```python

y_pred = regressor.predict(X_test) 
# Predicting 
df = pd.DataFrame({'Hours': [i[0] for i in X_test],'Actual Score': y_test, 'Predicted Score': y_pred})  
df
```

```python
# test with hr= 9.25
hours = 9.25
own_pred = regressor.predict([[hours]])
print("Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))

from sklearn import metrics  
print('MAE:', 
      metrics.mean_absolute_error(y_test, y_pred))  
```
